```python
import os
import yaml
import pandas as pd
import chardet #encoding-detection
```

yaml -> merge -> csv(file)

```python
main_folder = "C:\prj_stock_analysis\datas\original_data"
all_yaml_data = []

def detect_encoding(file_path):
    with open(file_path, 'rb') as f:
        result = chardet.detect(f.read(10000))
    return result['encoding']

for folder_path, subfolders, files in os.walk(main_folder):
    for yaml_file in files:
        if yaml_file.endswith(('.yaml', '.yml')):
            file_path = os.path.join(folder_path, yaml_file)
            try:
                encoding = detect_encoding(file_path)
                with open(file_path, 'r', encoding=encoding) as f:
                    data = yaml.safe_load(f)
                    if isinstance(data, list):
                        all_yaml_data.extend(data)
                    elif isinstance(data, dict):
                        all_yaml_data.append(data)
            except Exception as e:
                print(f"Error reading {file_path}: {e}")

# Save to CSV
if all_yaml_data:
    df = pd.DataFrame(all_yaml_data)
    output_path = os.path.join(r"C:\prj_stock_analysis\datas","aligned.csv")
    df.to_csv(output_path, index=False)
    print(f"saved and the link here {output_path}")
else:
    print("no valid yaml")
```

Data time column(modification) -> split columns

```python
df_copy['date']=pd.to_datetime(df_copy['date'])
```

```python
df_copy['Day']=df_copy['date'].dt.day
df_copy['Month']=df_copy['date'].dt.month
df_copy['Year']=df_copy['date'].dt.year
df_copy['Time'] = df_copy['date'].dt.time
```

```python
df_copy = df_copy.drop("month", axis=1)
df_copy = df_copy.drop("date", axis=1)
```

```python
df_copy.head()
```

Sorting -> grouping -> individual ticker csv's

```python
output_folder = r"C:\prj_stock_analysis\datas\aggregated_data_1"
os.makedirs(output_folder, exist_ok=True)


df_copy = df_copy.sort_values(by=['Month','Day'])


for ticker, group in df_copy.groupby('Ticker'):
    file_name = f"{ticker}.csv"
    output_path = os.path.join(output_folder, file_name)
    group.to_csv(output_path, index=False)

print("successful")
```
